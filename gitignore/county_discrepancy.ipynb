```python
#Function to check county discrepancy across all the states between two dataframes
def county_discrepancy(df1, df2, state):
    print(f'Counties in first and not in second:')
    print(set([c for c in df1['County_State'].values if c.endswith(state)]).difference(set([c for c in df2['County_State'].values if c.endswith(state)])))
    print()
    print(f'Counties in second and not in first:')
    print(set([c for c in df2['County_State'].values if c.endswith(state)]).difference(set([c for c in df1['County_State'].values if c.endswith(state)])))
    print()
```

```python
#Checking for a single state
county_discrepancy(ankita, jacqueline, 'ak')
```

```python
#Check for all the states between two dataframes
for state in states: 
    print(f'{state}')
    county_discrepancy(ankita, jacqueline, state)
```
